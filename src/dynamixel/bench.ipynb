```python
from pypot.dynamixel import DxlIO
from pypot.dynamixel.protocol.v1 import *

from glob import glob

ports = glob('/dev/ttyACM*')
assert len(ports) == 1

port = ports[0]
print('Connecting on port: {}'.format(port))
dxl_io = DxlIO(port)

poulpe_id = 42
```

```python
ping_packet = DxlPingPacket(poulpe_id)
read_data_packet = DxlReadDataPacket(poulpe_id, 45, 1)
write_data_packet = DxlWriteDataPacket(poulpe_id, 45, [1, 2, 3])
long_write_data_packet = DxlWriteDataPacket(poulpe_id, 12, list(range(50)))
```

```python
dxl_io._send_packet(ping_packet, wait_for_status_packet=True)
```

```python
%timeit dxl_io._send_packet(ping_packet, wait_for_status_packet=True)
```

```python
dxl_io._send_packet(read_data_packet, wait_for_status_packet=True)
```

```python
%timeit dxl_io._send_packet(read_data_packet, wait_for_status_packet=True)
```

```python
dxl_io._send_packet(write_data_packet, wait_for_status_packet=True)
```

```python
%timeit dxl_io._send_packet(write_data_packet, wait_for_status_packet=True)
```

```python
dxl_io._send_packet(long_write_data_packet, wait_for_status_packet=True)
```

```python
%timeit dxl_io._send_packet(long_write_data_packet, wait_for_status_packet=True)
```
